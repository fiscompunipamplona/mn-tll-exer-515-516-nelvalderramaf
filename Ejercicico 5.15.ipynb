```python
from math import tanh , cosh
from numpy import linspace 
from pylab import plot, show, xlabel, ylabel
```

Calculo de la derivada en forma exacta(Grafica Negra) y por derivadas centrales(Grafica Roja); comparacion y error promedio hallado entre estos. 

```python
df_dx=[]
df_dx_teo=[]
def f(x):
    return (1+((tanh(x*2))/2))
a=-2
b=2
paso=100
h=(b-a)/paso
error=0
puntos=linspace(a,b,paso)
for x in puntos:
    l=(1 / cosh(2*x))** 2
    df_dx_teo.append(l)
    
    m=(f(x+(h))-f(x-(h)))/(2*h)
    df_dx.append(m)
    
    error=error+abs(((m-l)/l)*100)

plot(puntos,df_dx,'or')
plot(puntos,df_dx_teo,'k')
xlabel('x')
ylabel('df/dx')
show()

print("error=",error/paso,"%")
```
